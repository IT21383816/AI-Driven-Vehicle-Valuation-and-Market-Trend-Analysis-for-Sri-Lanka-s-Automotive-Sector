```python
label_encoder = LabelEncoder()
data['name _num'] = label_encoder.fit_transform(data['name'])
data['fuel _num'] = label_encoder.fit_transform(data['fuel'])
data['seller_type _num'] = label_encoder.fit_transform(data['seller_type'])
data['transmission _num'] = label_encoder.fit_transform(data['transmission'])
data['owner _num'] = label_encoder.fit_transform(data['owner'])
```

```python
X = data[['name _num','year','km_driven','fuel _num','seller_type _num','transmission _num','owner _num']]
y = data['selling_price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=25)
```

```python
def model_score(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' | ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_score(lr)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_score(rf)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_score(dt)

from sklearn.linear_model import Ridge
ridge_regression = Ridge()
model_score(ridge_regression)
```

```python
import pickle

pickle.dump(rf, open("v_price_model.dat", "wb"))
```

```python
predict=np.round(rf.predict(X_test))
```

```python
A = np.array(y_test).reshape(-1, 1)
B = predict.reshape(-1, 1)
print(type(predict))
print(type(y_test))
plt.rcParams['figure.figsize'] = 16,5
plt.figure()
plt.plot(A[-100:], label="Real")
plt.plot(B[-100:], label="Predicted")
plt.legend()
plt.title("Result : real vs predicted")
plt.ylabel("Result")
plt.xticks(())
plt.show()
```
